```python
def res_generator():
    tests = {}
    trains = {}
    prds = {}

    for i in ['t', 'c', 'f', 'l']:
        print('cuurent item: {}'.format(i))
        train_all, test_all = Create_train_test(91, 121, i)

        d = [unsignificant(train_all), unsignificant(test_all)]
        intsection = list(set(d[0]).intersection(*d[:1]))
        train_all = train_all.drop(intsection, axis=1)
        test_all = test_all.drop(intsection, axis=1)
        X_train, Y_train = t_x_y(train_all)
        X_test, Y_test = t_x_y(test_all)
        score, ref = decicionTest(X_train, Y_train, X_test, Y_test)
        trains[i] = train_all
        tests[i] = test_all
        prds[i] = ref[0]

    # merge test & predict
    custPrds = {}
    for i in ['t', 'c', 'f', 'l']:
        tests[i]['{}_prediction'.format(i)] = prds[i]
        custPrds[i] = tests[i][['CUST_NO', '{}_prediction'.format(i)]]
    return custPrds, [tests, trains, prds]
```
